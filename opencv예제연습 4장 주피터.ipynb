```python
#예제 4.6 ROI에 의한 블록 평균 영상
import cv2
import numpy as np

src = cv2.imread('0205.png\cat.jpg', cv2.IMREAD_GRAYSCALE)
dst = np.zeros(src.shape, dtype = src.dtype)

N = 32
height, width = src.shape
##height, width, channel = src.shape

h = height // N
w = width // N
for i in range(N):
  for j in range(N):
    y = i * h
    x = j * w
    roi = src[y:y + h, x:x + w]
    dst[y:y + h, x:x + w] = cv2.mean(roi)[0]
##  dst[y:y + h, x:x + w] = cv2.mean(roi)[0:3]


cv2.imshow('dst', dst)   
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#예제 4.18 영상 연산 2: 비트연산
import cv2
import numpy as np

src1 = cv2.imread('0205.png\image_1.jpg')
src2 = cv2.imread('0205.png\logo.png')
cv2.imshow('src2', src2)

#1
rows,cols,channels = src2.shape
roi = src1[0:rows, 0:cols]

#2
gray = cv2.cvtColor(src2, cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(gray, 160, 255, cv2.THRESH_BINARY)
mask_inv = cv2.bitwise_not(mask)
cv2.imshow('mask', mask)
cv2.imshow('mask_inv', mask_inv)

#3
src1_bg = cv2.bitwise_and(roi, roi, mask = mask)
cv2.imshow('src1_bg', src1_bg)

#4
src2_fg = cv2.bitwise_and(src2, src2, mask = mask_inv)
cv2.imshow('src2_fg', src2_fg)

#5
##dst = cv2.add(src1_bg, src2_fg)
dst = cv2.bitwise_or(src1_bg, src2_fg)
cv2.imshow('dst', dst)

#6
src1[0:rows, 0:cols] = dst

cv2.imshow('result', src1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#예제 4.19 반전영상
import cv2
import numpy as np

src1 = cv2.imread('0205.png\image_1.jpg', cv2.IMREAD_GRAYSCALE)
src2 = np.zeros(shape = (700,900), dtype= np.uint8) + 255

dst1 = 255 - src1
dst2 = cv2.subtract(src2, src1)
dst3 = cv2.compare(dst1, dst2, cv2.CMP_NE)      # cv2.CMP_EQ
n    = cv2.countNonZero(dst3)
print('n = ', n)

cv2.imshow('dst1', dst1)
cv2.imshow('dst2', dst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#예제 4.20 cv2.normalize()에 의한 영상 정규화
import cv2
import numpy as np

src = cv2.imread('0205.png\image_1.jpg', cv2.IMREAD_GRAYSCALE)

minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(src)
print('src:', minVal, maxVal, minLoc, maxLoc)

dst = cv2.normalize(src, None, 100, 200, cv2.NORM_MINMAX)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(dst)
print('dst:', minVal, maxVal, minLoc, maxLoc)

cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#예제 4.21 cv2.randu()에 2차원 균등분포 난수 좌표
import cv2
import numpy as np
import time

dst = np.full((512, 512, 3), (255, 255, 255), dtype= np.uint8)
nPoints = 100
pts = np.zeros((1, nPoints, 2), dtype= np.uint16)
cv2.setRNGSeed(int(time.time()))
cv2.randu(pts, (0,0), (512, 512))

# draw points
for k in range(nPoints):
  x, y = pts[0, k][:]         #pts[0, k, :]
  cv2.circle(dst, (x,y), radius= 5, color= (0, 0, 255),
              thickness= -1)

cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#예제 4.22 cv2.randu()에 2차원 균등분포 난수 좌표
import cv2
import numpy as np
import time

dst = np.full((512, 512, 3), (255, 255, 255), dtype= np.uint8)
nPoints = 100
pts = np.zeros((1, nPoints, 2), dtype= np.uint16)

cv2.setRNGSeed(int(time.time()))
cv2.randn(pts, mean = (256,256), stddev = (50, 50))

# draw points
for k in range(nPoints):
  x, y = pts[0, k][:]         #pts[0, k, :]
  cv2.circle(dst, (x,y), radius= 5, 
            color= (0, 0, 255), thickness= -1)

cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
